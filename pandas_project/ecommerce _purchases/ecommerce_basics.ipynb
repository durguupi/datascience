## 14. How many people have a credit card that expires in 2020?

```python
data.head()
```

```python
data.info()
```

```python
data["CC Exp Date"] 
```

```python
def fun():
    count = 0
    for date in data["CC Exp Date"]:
        if date.split('/')[1] == "20":
            count = count + 1
    print(count)
        
```
