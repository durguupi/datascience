# DATA FLITERING

```python
df["Score"] > 7
```

```python
# To filter the data first we have to create
# 1. Create boolean mask . Its an array of [True, False]
# Assign it to variable
# Then get the data from it
greater_than_7_boolean_mask = df["Score"] >= 7
```

```python
df[greater_than_7_boolean_mask]
```

```python
# Filter by multiple conditions using AND /Or Operators
df["Healthy life expectancy"] < 1
```

```python
life_expectancy_less_than_one = df["Healthy life expectancy"] < 1
```

```python
df[life_expectancy_less_than_one]
```

```python
(df["Score"] >= 7) & (df["Healthy life expectancy"] < 1)
```

```python
score_over_7_life_exp_less_than_one = (df["Score"] >= 7) & (df["Healthy life expectancy"] < 1)
```

```python
df[score_over_7_life_exp_less_than_one]
```

```python
# Different way of expressing it 
df[
    (df["Score"] >= 7) & 
    (df["Healthy life expectancy"] < 1)
]
```

# USE OF MAP FUNCTION

```python
# PASS a function that accepts a input and gives output
def map_score_to_category(score):
    if score >= 7:
        return "High Score"
    elif score < 4:
        return "Low Score"
    else:
        return "Medium Score"
    
```

```python
df["Score"].map(map_score_to_category)
```

```python
df["Score"].map(map_score_to_category).value_counts()
```

```python
df["Score_caetgory"] = df["Score"].map(map_score_to_category)
```

```python
df.head()
```
